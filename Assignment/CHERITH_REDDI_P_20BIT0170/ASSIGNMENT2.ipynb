```python
data['age']=data['age'].fillna(data['age'].mean())
```

```python
data.isnull().sum()
```

```python
data["deck"]=data["deck"].fillna(data["deck"].mode().iloc[0])
```

```python
# #as embarked,embark_town nan values are too les we dop it 
# data['embark_town'].dropna(inplace=True)
# data['embarked'].dropna(inplace=True)
data.dropna(subset=['embark_town', 'embarked'], inplace=True)
```

```python
data.isnull().sum()
```

6. Find the outliers and replace the outlier

```python
columns=['pclass','age','sibsp','parch','fare']
for column in columns:
  q1=data[column].quantile(0.25)
  q3=data[column].quantile(0.75)
  IQR=q3-q1
  lowerbound=q1-(1.5*IQR)
  upperbound=q3+(1.5*IQR)
  # print(column,q1,q3,IQR,lowerbound,upperbound)


  #outliers
  outliers = data[(data[column] < lowerbound) | (data[column] > upperbound)]
  print(column,outliers)
  #replacing with median
  data.loc[(data[column] < lowerbound) | (data[column] > upperbound), column] = data[column].median()



```
